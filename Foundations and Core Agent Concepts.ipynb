<h2 style="color:#28B463;">Agent Anatomy</h2>

<p>
Each agent consists of several internal components that enable intelligent behavior:
</p>

<h3 style="color:#AF7AC5;">Goals</h3>
<p>
the objectives or desired outcomes the agent seeks to achieve, which may be updated based on feedback or changing context.</p>

<h3 style="color:#AF7AC5;">Sense (Perception)</h3>
<p>
This component is how the agent gathers information and data from its environment (digital or physical sources such as APIs, databases, and sensors). This perception process is the agent's mechanism for acquiring context, the situational awareness that is critical for all subsequent reasoning and decision-making, reinforcing the principle that "context is king." One popular mechanism to standardize how models access this contextual information is through protocols such as the Model Context Protocol (MCP) from Anthropic (http://anthropic.ai/).</p>


<h3 style="color:#AF7AC5;">Reason (thinking models, cognition) </h3>
<p>
The core processing unit where sensed information is analyzed. This often heavily involves LLMs for interpreting data, understanding relationships (between goals, perception, and actions), and complex inference.
</p>

<ul>
<li>Data is analyzed</li>
<li>Relationships are understood</li>
<li>Complex reasoning and inference occur (often using LLMs)</li>
</ul>

<h3 style="color:#AF7AC5;">Plan</h3>
<p>
Devising a course of action or sequence of steps based on the reasoned insights and current goals.</p>

<h3 style="color:#AF7AC5;">Act (Action)</h3>
<p>
Executing the planned actions upon the environment using available tools (e.g., calling APIs, controlling robotic elements, and generating text).
</p>

<ul>
<li>API calls</li>
<li>System operations</li>
<li>Content generation</li>
<li>Physical control (robots, devices)</li>
</ul>

<h3 style="color:#AF7AC5;">Memory</h3>
<p>
Stores: Stores the agent's individual knowledge, past experiences, internal state, and learned information, providing context for decision-making.
</p>

<ul>
<li>Past experiences</li>
<li>Learned knowledge</li>
<li>Internal state</li>
</ul>

<p>
Memory enables:
</p>

<ul>
<li>Context retention</li>
<li>Better future decision-making</li>
</ul>

<h3 style="color:#AF7AC5;">Coordinate (Multi-Agent Only)</h3>
<p>
Allows agents to collaborate with others through:

 Interacting with other agents, often via shared memory or communication protocols, to align actions and collaborate toward collective goals. This may involve negotiation or following specific protocols. This coordination is specifically recommended to occur via the agent-to-agent (A2A) interoperability protocol.
The agent's ability to act is typically enabled by a mechanism known as function calling. To instruct the LLM, developers provide it with a list of available tools. Each tool is defined with a name, a clear description of its purpose, and a structured schema of its required parameters. Based on the ongoing task, the LLM's reasoning core decides when to use a tool, which tool is most appropriate, and what parameters to use. The model then generates a structured output, such as a JSON object, signaling its intent to call that function with the extracted arguments. The agent's code receives this, executes the function, and feeds the result back to the LLM to continue its operational loop.

</p>

<ul>
<li>Shared memory</li>
<li>Communication protocols</li>
<li>Negotiation and coordination strategies</li>
</ul>

<p>
Standardization example:
</p>
<ul>
<li>Agent-to-Agent (A2A) protocol</li>
</ul>

<h2 style="color:#E67E22;">Function Calling (Tool Usage)</h2>

<p>
Agents interact with external systems using <b>function calling</b>.
</p>

<p><b>How it works:</b></p>

<ul>
<li>Developers define tools with:
  <ul>
    <li>Name</li>
    <li>Description</li>
    <li>Parameter schema</li>
  </ul>
</li>
<li>LLM decides:
  <ul>
    <li>When to use a tool</li>
    <li>Which tool to use</li>
    <li>What parameters to pass</li>
  </ul>
</li>
<li>Model generates structured output (e.g., JSON)</li>
<li>System executes the function</li>
<li>Results are fed back to the agent</li>
</ul>


<h1 style="color:#2E86C1;">7. The New Agentic Stack</h1>

<p>
As AI systems evolve from simple prompt-based interactions to <b>agentic architectures</b>, the ability for models and agents to interact reliably with tools and each other becomes critical.
</p>

<p>
This evolution introduces a new interoperability stack consisting of:
</p>

<ul>
<li><b>Function Calling</b></li>
<li><b>Model Context Protocol (MCP)</b></li>
<li><b>Agent-to-Agent (A2A) Protocol</b></li>
</ul>

<hr>

<h2 style="color:#28B463;">Function Calling (Tool Invocation)</h2>

<p>
Function calling allows LLMs within an agent to intelligently invoke external tools.
</p>

<p><b>Examples:</b></p>
<ul>
<li><code>book_flight(destination="Tokyo")</code></li>
<li><code>get_credit_score()</code></li>
<li>Execute Python scripts for data analysis</li>
</ul>

<p>
This enables agents to move beyond text generation and <b>take real-world actions</b>.
</p>

<hr>

<h2 style="color:#AF7AC5;">Model Context Protocol (MCP)</h2>

<p>
MCP provides a standardized way to:
</p>

<ul>
<li>Describe tools</li>
<li>Discover available services</li>
<li>Securely invoke external systems</li>
</ul>

<p><b>Examples of MCP-enabled tools:</b></p>
<ul>
<li>Weather APIs</li>
<li>Vector search systems</li>
<li>Enterprise APIs (e.g., property verification)</li>
</ul>

<p>
<b>Key Benefit:</b> Enhances modularity and interoperability by treating tools as independent services.
</p>

<hr>

<h2 style="color:#E67E22;">Agent-to-Agent (A2A) Protocol</h2>

<p>
A2A enables structured communication between multiple agents.
</p>

<ul>
<li>Task delegation</li>
<li>Collaboration</li>
<li>Coordination and negotiation</li>
</ul>

<p>
This is essential for <b>multi-agent systems (Level 6)</b>.
</p>

<hr>

<h2 style="color:#1F618D;">MCP vs A2A (Key Distinction)</h2>

<ul>
<li><b>MCP → Vertical Integration</b><br>
Agent connects to tools, data, and external systems</li>

<li><b>A2A → Horizontal Integration</b><br>
Agents connect and collaborate with other agents</li>
</ul>

<p><b>Simple Mental Model:</b></p>

<ul>
<li>MCP = AI agent ↔ Tools</li>
<li>A2A = AI agents ↔ AI agents</li>
</ul>

<hr>

<h2 style="color:#7D3C98;">How They Work Together</h2>

<p>
These protocols are complementary and often used together:
</p>

<ul>
<li>An <b>orchestrator agent</b> uses A2A to delegate tasks</li>
<li>Worker agents use MCP to access tools and data</li>
<li>Results flow back through A2A</li>
</ul>

<p>
This creates a <b>collaborative, scalable workflow</b>.
</p>

<hr>

<h2 style="color:#117864;">Multi-Agent Architecture (Conceptual)</h2>

<p>
In a distributed system:
</p>

<ul>
<li>Each agent has its own:
  <ul>
    <li>LLM orchestration</li>
    <li>Memory</li>
    <li>Toolchain</li>
  </ul>
</li>

<li>Agents access external tools via <b>MCP</b></li>
<li>Agents communicate with each other via <b>A2A</b></li>
</ul>

<p>
This enables independent yet coordinated operation across the system.
</p>

<hr>

<h2 style="color:#C0392B;">Key Takeaway</h2>

<p>
The agentic stack enables <b>modular, scalable, and collaborative AI systems</b> by combining:
</p>

<ul>
<li>Tool interaction (Function Calling)</li>
<li>Tool standardization (MCP)</li>
<li>Agent collaboration (A2A)</li>
</ul>

<p>
Mastering this stack is essential for building <b>advanced agentic AI systems</b>.

<h1 style="color:#2E86C1;">Real-World System Design: Agentic Loan Processing System</h1>

<p>
We design a multi-agent system for automating loan underwriting in a financial institution.
</p>

<p><b>Goal:</b> Automate decision-making while ensuring accuracy, compliance, and scalability.</p>

<h2 style="color:#28B463;">System Architecture</h2>

<ul>
<li><b>Orchestrator Agent</b> – Controls workflow</li>
<li><b>Document Agent</b> – Validates documents</li>
<li><b>Credit Agent</b> – Fetches credit score</li>
<li><b>Risk Agent</b> – Evaluates risk</li>
</ul>

<p><b>External Systems:</b></p>
<ul>
<li>Credit Bureau API</li>
<li>Document DB</li>
<li>Policy Knowledge Base (RAG)</li>
</ul>

<p><b>Protocols:</b></p>
<ul>
<li>MCP → Tool access</li>
<li>A2A → Agent communication</li>
</ul>

<h2 style="color:#AF7AC5;">Workflow</h2>

<ol>
<li>User submits loan application</li>
<li>Orchestrator decomposes tasks</li>
<li>Agents execute tasks in sequence</li>
<li>Results aggregated</li>
<li>Final decision returned</li>
</ol>

```python
# Install & Setup

!pip install openai
```

```python

## Define Tool Functions (MCP-like layer)
# Tools simulate external APIs

def verify_documents(applicant):
    return {"status": "verified"}

def get_credit_score(applicant):
    return {"credit_score": 720}

def risk_assessment(data):
    if data["credit_score"] > 700:
        return {"risk": "low", "decision": "approve"}
    return {"risk": "high", "decision": "reject"}
```

```python
## Base Agent Class
class Agent:
    def __init__(self, name):
        self.name = name

    def act(self, input_data):
        raise NotImplementedError
```

```python
## Specialized Agents
class DocumentAgent(Agent):
    def act(self, applicant):
        return verify_documents(applicant)

class CreditAgent(Agent):
    def act(self, applicant):
        return get_credit_score(applicant)

class RiskAgent(Agent):
    def act(self, data):
        return risk_assessment(data)
```

```python
## Orchestrator Agent
class LoanOrchestratorAgent(Agent):
    def __init__(self):
        super().__init__("Orchestrator")
        self.doc_agent = DocumentAgent("DocAgent")
        self.credit_agent = CreditAgent("CreditAgent")
        self.risk_agent = RiskAgent("RiskAgent")

    def act(self, applicant):
        print("Step 1: Document Verification")
        doc_result = self.doc_agent.act(applicant)

        print("Step 2: Credit Check")
        credit_result = self.credit_agent.act(applicant)

        combined_data = {**doc_result, **credit_result}

        print("Step 3: Risk Assessment")
        risk_result = self.risk_agent.act(combined_data)

        return risk_result
```

```python
## Run the System
orchestrator = LoanOrchestratorAgent()

applicant = {
    "name": "John Doe",
    "income": 50000
}

result = orchestrator.act(applicant)

print("\nFinal Decision:", result)
```
